```python
from minepy import MINE
X_mic = []
mine = MINE(alpha = 0.6, c = 15)
for i in X:
    mine.compute_score(i, y)
    if mine.mic() >= 0.2:
        X_mic.append(i)

X_mic = np.array(X_mic)
nrow = X_mic.shape[0]
ncol = X_mic.shape[1]
X_mic_new = []
X_mic_new_one = []

for i in range(0, ncol):
    for j in range(0, nrow):
        X_mic_new_one.append(X_mic[j][i])
    X_mic_new.append(X_mic_new_one)
    X_mic_new_one = []
X_mic_new = np.array(X_mic_new)
X_mic_new.shape
```
