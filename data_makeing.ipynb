```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([8, 10, 10, 11, 13, 15, 11, 17, 16, 15, 14, 16, 13, 14, 11, 12])
y = []
x_ax = np.arange(len(x))
for i in x:
    y.append(32 - i)
y = np.array(y)
plt.style.use("default")
plt.xlabel("周数")
plt.ylabel("人数")
plt.bar(x_ax - 0.175, x, width = 0.35, label = "高压组")
plt.bar(x_ax + 0.175, y, width = 0.35, label = "低压组")
plt.legend()
```
